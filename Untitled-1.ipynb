```python
import numpy as np
import pandas as pd
import os
```

```python
path_data_train = os.path.join(os.getcwd(), 'data', 'train.csv')
print(path_data_train)
```

```python
df = pd.read_csv(path_data_train)
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
categorical_colums = df.dtypes[df.dtypes == 'object'].index
categorical_colums = categorical_colums[1:].to_list()
```

```python
df.shape
```

```python
for collum in df.columns:
    print(collum, df[collum].nunique())
```

```python
import matplotlib.pyplot as plt

# ,  DataFrame  df
for column in categorical_colums:
    #       
    counts = df[column].value_counts()
    
    #   
    plt.figure(figsize=(8, 6))
    plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=140)
    plt.title(f'   : {column}')
    plt.show()
```

```python
numerical_desc = df.describe()
print(numerical_desc)

# Summary of categorical columns
categorical_desc = df.describe(include=['object'])
categorical_desc
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print(missing_values)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histograms for numerical columns
numerical_cols = ['Item_Weight', 'Item_Visibility', 'Item_MRP', 'Item_Outlet_Sales']

for col in numerical_cols:
    plt.figure(figsize=(10, 6))
    sns.histplot(df[col], bins=30, kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()

# Bar plots for categorical columns
categorical_cols = ['Item_Fat_Content', 'Item_Type', 'Outlet_Size', 'Outlet_Location_Type', 'Outlet_Type']

for col in categorical_cols:
    plt.figure(figsize=(10, 6))
    sns.countplot(y=df[col])
    plt.title(f'Count of {col}')
    plt.show()
```

```python
# Correlation matrix for numerical columns
correlation_matrix = df[numerical_cols].corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Create a new feature for the age of the outlet
current_year = 2024  # Replace with the current year
df['Outlet_Age'] = current_year - df['Outlet_Establishment_Year']
print(df[['Outlet_Establishment_Year', 'Outlet_Age']].head())
```

```python
# Sales by Item_Type
plt.figure(figsize=(14, 8))
sns.boxplot(x='Item_Type', y='Item_Outlet_Sales', data=df)
plt.xticks(rotation=90)
plt.title('Sales by Item Type')
plt.show()

# Sales by Outlet_Type
plt.figure(figsize=(10, 6))
sns.boxplot(x='Outlet_Type', y='Item_Outlet_Sales', data=df)
plt.title('Sales by Outlet Type')
plt.show()

# Sales by Outlet_Location_Type
plt.figure(figsize=(10, 6))
sns.boxplot(x='Outlet_Location_Type', y='Item_Outlet_Sales', data=df)
plt.title('Sales by Outlet Location Type')
plt.show()
```
